```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
 
```

```python
url='https://www.flipkart.com/search?q=mobile&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off'
html_content=requests.get(url)
data=html_content.content
```

```python
data_file=bs(data,'html.parser')
```

```python
data_file
data_file.prettify
```

```python
mobile_name=data_file.find_all("div", class_='_4rR01T')
list_name=[]
for name in mobile_name:
  a=name.get_text()
  list_name.append(a)
list_name
```

```python
mobile_price=data_file.find_all("div", class_='_30jeq3 _1_WHN1')
list_price=[]
for price in mobile_price:
  a=price.get_text()
  list_price.append(a)
list_price
```

```python
dict={
    'mobile_name':list_name,
    'mobile_price':list_price
}
```

```python
df=pd.DataFrame(dict)
df.head(5)
```

```python
df.to_csv("Mobile")
```
